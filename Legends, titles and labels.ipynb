```python
import matplotlib.pyplot as plt
#legends differentiate, they are like a key.
```

```python
#plotting from lists.
x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]
```

```python
plt.plot(x,y, label="First Line",c='red')
plt.plot(x2,y2, label ="Second Line",c='blue')
plt.xlabel('x labels')
plt.ylabel('y labels')
plt.title('Basic Interesting Graph\nCheck it out')
plt.legend()
plt.show()
#Discussion customization later.
```
