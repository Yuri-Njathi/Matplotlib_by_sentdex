```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
y = [2,4,2,3,1,7,3,8,2]

x2 = [6,7,8,9,1,2,3,4,5]
y2 = [1,7,3,8,2,2,4,2,3]

plt.scatter(x,y,label='skitscat',color='k',marker='x',s=100)
plt.scatter(x2,y2,label='scat',color='c',marker='*',s=100)
#s is for size
#markers search on google matplotlib marker options

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it Out')
plt.legend()
plt.show()
```
